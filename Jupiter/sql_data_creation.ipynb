```python
df=pd.read_csv(r"C:\Users\acer\Downloads\Sample - Superstore.csv", encoding='latin1')  
df
```

```python
df1=df.copy()
df1
```

```python
df1.columns=df.columns.str.strip().str.lower().str.replace(' ','_')
```

```python
df1.info()
```

```python
pd.set_option('display.max.columns',27)
df1
```

```python
df2=df1.copy()
df2
```

```python
#creating unique product_id by combining category,subcategory and productname

df2['product_id']=df2['category']+"_"+df2['sub-category']+ "_" +df2['product_name']

#creating unique sale_id by combining order_id and index
df2['sale_id']=df2['order_id']+"_"+df2.index.astype(str)
```

```python
df2
```

```python
#1 Customers table
customers=df2[['customer_id','customer_name','segment','country','city','state','postal_code','region']].drop_duplicates()
customers.columns=['customer_id','name','segment','country','city','state','postal_code','region']


# Products table
products=df2[['product_id','category','sub-category','product_name']].drop_duplicates()
products.columns=['product_id','category','sub_category','product_name']

# Sales table
sales=df2[['sale_id','order_id','customer_id','product_id','order_date','ship_date','ship_mode','sales','quantity','discount','profit']]
```

```python
pd.set_option('display.max.rows',4910)
customers
products
sales
```

```python
user="root"
password="predestination"
host="localhost"
port=3306
database="superstore"

#connect to mysql using sqlalchemy
engine=create_engine(f"mysql+mysqlconnector://{user}:{password}@{host}:{port}/{database}")

#test the connection
with engine.connect() as conn:
    result= conn.execute(text("SHOW TABLES;"))
    print("Tables in database:", [row[0] for row in result])
```

```python
#Insert data into mysql tables from pandas

customers.to_sql("customers",con=engine,if_exists="replace",index=False)
products.to_sql("products",con=engine,if_exists="replace",index=False)
sales.to_sql("sales",con=engine,if_exists="replace",index=False)
```
